```python
import nltk
from nltk.stem.lancaster import LancasterStemmer
import os
import json
import datetime
import pandas as pd

stemmer = LancasterStemmer()
df = pd.read_csv("~/desktop/machinelearning/trainmini.csv",encoding='latin-1').fillna("")
from nltk.corpus import stopwords
from string import digits
import string
import re
words = []
classes = []
document1 = []
document2 = []
ignore_words = ['?']
# loop through each sentence in our training data
for index, row in df.iterrows():
    row['question1'].translate(str.maketrans('','',string.punctuation))
    row['question2'].translate(str.maketrans('','',string.punctuation))

    row['question1'] = re.sub('[!\[\]@#\\\.?,$/]', '', row['question1'])
    row['question2'] = re.sub('[!\[\]@#\\\.?,$/]', '', row['question2'])
    remove_digits = str.maketrans('', '', digits)
    row['question1'] = row['question1'].translate(remove_digits)
    row['question2'] = row['question2'].translate(remove_digits)
    w = nltk.word_tokenize(row['question1'])
    w2 = nltk.word_tokenize(row['question2'])
    # add to our words list
    words.extend(w)
    words.extend(w2)
    # add to documents in our corpus
    document1.append((w, row['is_duplicate']))
    document2.append((w2, row['is_duplicate']))
    # add to our classes list
    if row['is_duplicate'] not in classes:
        classes.append(row['is_duplicate'])

# stem and lower each word and remove duplicates
words = [stemmer.stem(w.lower()) for w in words if w not in stopwords.words('english')]

words = list(set(words))


print (len(document1), "document1")
print (len(document2), "document2")
print (len(classes), "classes", classes)
print (len(words), "unique stemmed words", words)

# create our training data
training = []
output = []
# create an empty array for our output
output_empty = [0] * len(classes)

# training set, bag of words for each sentence
for doc in document1:
    # initialize our bag of words
    bag = []
    # list of tokenized words for the pattern
    pattern_words = doc[0]
    # stem each word
    pattern_words = [stemmer.stem(word.lower()) for word in pattern_words]
    # create our bag of words array
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)

    training.append(bag)
    # output is a '0' for each tag and '1' for current tag
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1
    output.append(output_row)
for doc in document2:
    # initialize our bag of words
    bag = []
    # list of tokenized words for the pattern
    pattern_words = doc[0]
    # stem each word
    pattern_words = [stemmer.stem(word.lower()) for word in pattern_words]
    # create our bag of words array
    for w in words:
        bag.append(1) if w in pattern_words else bag.append(0)

    training.append(bag)
    # output is a '0' for each tag and '1' for current tag
    output_row = list(output_empty)
    output_row[classes.index(doc[1])] = 1
    output.append(output_row)
# sample training/output
i = 0
w0 = document1[i][0]
w1 = document2[i][0]
print ([stemmer.stem(word.lower()) for word in w])
print (training[i])
print (output[i])
for a,b in zip(document1,document2):
    print (a)
    print (a[0])
    print (b)
    print (b[0])
    
from sklearn.feature_extraction.text import TfidfVectorizer, HashingVectorizer, CountVectorizer
corpus = []
for a,b in zip(document1,document2):
    print (a)
    print (a[0])
    print (b)
    print (b[0])
    corpus.append(a[0])
    corpus.append(b[0])
    aVector = TfidfVectorizer(max_features = 10).fit_transform(a[0])
    bVector = TfidfVectorizer(max_features = 10).fit_transform(b[0])

vectorizer = CountVectorizer(min_df=1) 
analyze = vectorizer.build_analyzer()
print(corpus[0])
print(corpus[1])
newCorpus = []
corpus[0]=' '.join(corpus[0]).lower()
corpus[1]=' '.join(corpus[1]).lower()
newCorpus.append(corpus[0])
newCorpus.append(corpus[1])
print (newCorpus)
X = vectorizer.fit_transform(newCorpus)

vectorizer.get_feature_names()

minidoc1 = document1[:20]
minidoc2 = document2[:20]
import numpy as np
from nltk.stem.snowball import SnowballStemmer
def cosine(a0,a1):
    dotp = np.dot(a0, a1)
    mag0 = np.linalg.norm(a0)
    mag1 = np.linalg.norm(a1)
    cosineSim = dotp / (mag0* mag1)
    print(cosineSim)
    return cosineSim

def words_in_string(word_list, a_string):
    return set(word_list).intersection(a_string.split())

```
